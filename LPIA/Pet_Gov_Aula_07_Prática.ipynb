# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

Podemos agora criar um array para preencher as posições x e y de um gráfico

Exemplo:


```python
x = np.array([0, 6])
y = np.array([0, 250])
plt.plot(x, y)
plt.show()
```

Se desejamos plotar apenas os marcadores (sem a linha) podemos usar o parâmetro de notação de string "o" que significa um marcador do tipo anel; 

Exemplo:


```python
matriz_x = np.array([1, 8])
matriz_y = np.array([3, 10])
plt.plot(matriz_x, matriz_y, 'o')
plt.show()
```

Nós podemos ter quantos pontos quisermos e desejarmos no nosso gráfico. Entretanto, cabe lembrar que o número de pontos de x precisa ser exatamente o mesmo número de pontos de y. 

Exemplo:


```python
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])
plt.plot(x, y)
plt.show()
```

Com a finalidade de melhorar a interpretação e aparência gráfica nós podemos usar a palavra-chave marker para definir o tipo do marcador do gráfico.

Exemplo:


```python
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 'o')
plt.show()
```

Com a finalidade de melhorar a interpretação e aparência gráfica nós podemos usar a palavra-chave marker para definir o tipo do marcador do gráfico.

Exemplo:


```python
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = '*')
plt.show()
```

```python
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 'v')
plt.show()
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 'p')
plt.show()
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 'D')
plt.show()
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 's')
plt.show()
```

Através do mesmo parâmetro de notação de string nós podemos especificar o tipo da linha e a cor dos marcadores 

**marcador|linha|cor**

Exemplo:


```python
y = np.array([3, 8, 1, 10])
plt.plot(y, 'o:r')
plt.show()
```

As linhas podem ser dos seguintes tipos:

Exemplo:


```python
y = np.array([3, 8, 1, 10])
plt.plot(y, 'o-.r')
plt.show()
y = np.array([3, 8, 1, 10])
plt.plot(y, 'o--g')
plt.show()
y = np.array([3, 8, 1, 10])
plt.plot(y, 'o-b')
plt.show()
```

Também podemos alterar o tamanho dos marcadores usando a palavra-chave ms:

Exemplo:


```python
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 'o', 
ms = 20)
plt.show()
```

Também podemos alterar a cor da borda do marcador usando a palavra-chave mec:

Exemplo:


```python
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 'o', 
ms = 20, mec = 'r')
plt.show()
```

```python
y = np.array([3, 8, 1, 10])
plt.plot(y, marker = 'o', 
ms = 20, mfc = 'r')
plt.show()
```

Para alterar o estilo das linhas separadamente do marcador, podemos usar a palavra-chave *linestyle* ou ls:

Exemplo:


```python
y = np.array([3, 8, 1, 10])
plt.plot(y, ls = 'dotted')
plt.show()
```

Para alterar o estilo das linhas separadamente do marcador, podemos usar a palavra-chave linestyle ou ls:

Exemplo:


```python
y = np.array([3, 8, 1, 10])
plt.plot(y, ls = 'dashed')
plt.show()
```

Também podemos alterar a espessura da linha:

Exemplo:


```python
y = np.array([3, 8, 1, 10])
plt.plot(y, 
linewidth = '20.5')
plt.show()
```

Podemos ter várias linhas em um mesmo gráfico? Sim! 

Exemplo:


```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
plt.plot(y1)
plt.plot(y2)
plt.show()
```

Também podemos adicionar títulos e rótulos dos eixos 

Exemplo:


```python
x = np.array([80, 85, 90, 95, 100, 
105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 
280, 290, 300, 310, 320, 330])
plt.plot(x, y)
plt.title("Dados do smartwatch")
plt.xlabel("Média de pulsação")
plt.ylabel("Queima de Calorias")
plt.show()
```

Para fazermos um gráfico de dispersão (scatter) podemos usar este método  

Exemplo:

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)
plt.show()
```

Para fazermos um gráfico de barras (bar()) podemos usar este método

Exemplo:


```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, color = "red")
plt.show()
```

A vantagem do Seaborn é que nós podemos carregar diretamente um data set para ela

Lembre-se que precisamos ter as bibliotecas importadas:


```python
import matplotlib.pyplot as plt
import seaborn as sns


#vamos carregar um DataFrame
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
arquivo = '/content/drive/MyDrive/Forma.AI/material de apoio (colab)/datasets/aula_04_iris.csv'
df = pd.read_csv(arquivo) #você só terá acesso ao arquivo se autorizar o Colab a acessar o seu Google drive rodando o comando da célular anterior

sns.set_theme(style="ticks")
sns.pairplot(df, hue="class")
```

Agora, vamos visualizar a nossa árvore de decisão que geramos na aula anterior

```python
df = df[df['class'] != 'Iris-virginica'] #transformando em um problema binário (com apenas 2 rótulos de classe possíveis)
df.loc[df['class'] == 'Iris-setosa', 'class'] = 0
df.loc[df['class'] == 'Iris-versicolor', 'class'] = 1
df['class']=df['class'].astype('int')
X = np.array(df.drop(labels=['class'], axis=1))
y = np.array(df['class'])
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import plot_confusion_matrix
from sklearn import tree
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
from sklearn.tree import DecisionTreeClassifier
clf_dt = DecisionTreeClassifier(
    criterion='gini', splitter='best', max_depth=None,
    min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0,
    max_features=None, random_state=None, max_leaf_nodes=None,
    min_impurity_decrease=0.0, class_weight=None, ccp_alpha=0.0
)
clf_dt.fit(X_train, y_train)
y_pred = clf_dt.predict(X_test)
print('Classificações realizadas com sucesso')
```

```python
plt.rcParams['figure.figsize'] = (160, 70)
tree.plot_tree(
    clf_dt, # O modelo treinado
    fontsize=50, # Tamanho da fonte
    feature_names=df.drop(labels=['class'], axis=1).columns.to_list(), # Nomes das colunas (features) com exceção da coluna alvo
    class_names=['Iris-setosa', 'Iris-versicolor'], # Possíveis valores para a coluna alvo
    filled=True # Aplicar cores aos nós
)
```
