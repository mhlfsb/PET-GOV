# **K-mean**


## 1. O que K-Means faz?

Agrupamento K-means (K-means clustering) √© um tipo de aprendizado n√£o supervisionado utilizado quando temos dados n√£o rotulados (por exemplo, dados sem grupos ou categorias definidas). O objetivo deste algoritmo √© encontrar grupos no conjunto de dados, sendo o n√∫mero de grupos formado representado pela vari√°vel K. O algoritmo trabalha iterativamente para atribuir cada um dos dados a um dos K grupos com base nas caracter√≠sticas fornecidas.  Os dados s√£o agrupados por similaridade. Os resultados gerados pelo algoritmo Agrupamento K-means (K-means clustering) s√£o:

* Os centroides dos K clusters (agrupamentos), que podem ser usados para rotular novos dados 
* R√≥tulos para os dados de treinamento (cada um dos dados √© atribu√≠do a um √∫nico cluster) 



## 2. Aplica√ß√µes </br>
O algoritmo Agrupamento K-means (K-means clustering) √© usado para encontrar grupos que n√£o tenham sido explicitamente rotulados no conjunto de dados. Pode ser usado para confirmar suposi√ß√µes sobre que tipos de grupos existem ou identificar grupos desconhecidos em conjuntos de dados complexos.  Uma vez que o algoritmo tenha sido executado e os grupos definidos, qualquer novo dado pode ser facilmente atribu√≠do ao grupo correto.

√â um algoritmo vers√°til que pode ser usado para qualquer tipo de agrupamento. Alguns exemplos de casos de uso s√£o:

* Segmenta√ß√£o comportamental:
** Segmento por hist√≥rico de compras
** Segmento por atividades em aplicativo, website ou plataforma
* * Defini√ß√£o de perfils baseados em interesses 
* * Cria√ß√£o de perfis com base em monitora√ß√£o de atividade
* Categoriza√ß√£o de Invent√°rio:
* * Invent√°rio de grupo por atividade de vendas
* * Invent√°rio de grupo por m√©tricas de fabrica√ß√£o
* Classifica√ß√£o de medidas de sensores:
* * Detec√ß√£o de tipos de atividade em sensores de movimento
* * Agrupar imagens
* * Separar √°udio
* * Identificar grupos no monitoramento de sa√∫de
* Detectar rob√¥s ou anomalias:
* * Separar grupos de atividades v√°lidos de rob√¥s

**Inicializa√ß√£o com K-mean++** 


O algoritmo K-mean++ garante uma inicializa√ß√£o mais inteligente dos centroides e melhora a qualidade do agrupamento. Al√©m da inicializa√ß√£o, o resto do algoritmo √© igual ao K-means padr√£o. 

Algoritmo de inicializa√ß√£o:


1.   Selecione aleatoriamente o primeiro centroide entre os dados existentes.
2.   Para cada dado (observa√ß√£o), calcule sua dist√¢ncia do centr√≥ide mais pr√≥ximo, previamente escolhido.
3.   Selecione o pr√≥ximo centr√≥ide entre as observa√ß√µes (dados) de forma que a probabilidade de escolher uma observa√ß√£o como centr√≥ide seja diretamente proporcional √† sua dist√¢ncia do centr√≥ide mais pr√≥ximo. Ou seja, observa√ß√£o com dist√¢ncia m√°xima do centr√≥ide mais pr√≥ximo provavelmente ser√° selecionado como centr√≥ide.
4.   Repita os passos 2 e 3 at√© que k centroides tenham sido amostrados

## 4. Escolhendo o K correto

Uma forma de escolher o K correto √© utilizando-se de um par√¢metro conhecido como WCSS que √© a sigla para Soma dos Quadrados Dentro de um Grupo (**Within Cluster Sum of Squares**). Aqui est√° a f√≥rmula para quando o K=3

image.png

Quanto menor o WCSS, mais pr√≥ximo cada observa√ß√£o est√° do seu respectivo centr√≥ide. Em outras palavras, quanto menor WCSS, mais homog√™neo √© cada cada cluster. 

### Elbow Method



O M√©todo do Cotovelo (Elbow Method) pode ser utilizado para escolher o melhor valor de K. **O m√©todo consiste em plotar o WCSS em fun√ß√£o do n√∫mero de clusters**. O gr√°fico tem um formato de cotovelo e **geralmente** √© uma boa ideia escolher o n√∫mero onde esse cotovelo se forma. </br>**Obs.: Na pratica haver√° muitos casos em que o gr√°fico n√£o ser√° t√£o intuitivo.**

### Elbow Method

Vamos utilizar o K-means da biblioteca Scikit-learn (sklearn) </br>
https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

```python
# Utilizando o m√©todo do cotovelo para encontrar o n√∫mero ideal de grupos
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42) #Utilizamos o m√©todo de inicializa√ß√£o k-mean++
    kmeans.fit(X)
    # inertia method returns wcss for that model
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(range(1, 11), wcss,marker='o',color='red')
plt.title('Elbow Method')
plt.xlabel('N√∫mero de Grupos')
plt.ylabel('WCSS')
plt.show()
```

De acordo com o m√©todo do cotovelo devemos escolher K = 5! </br>
Portanto, vamos treinar nosso modelo com o parametro `n_clusters = 5`

```python
# Fitting K-Means to the dataset
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

## M√©tricas 
 
* Erro de Quantiza√ß√£o: √â a dist√¢ncia m√©dia entre uma observa√ß√£o e o n√≥ que melhor a representa (Best Matching Unit). √â utilizado como m√©trica b√°sica para avaliar SOM.
* Erro Topogr√°fico: √â a medida de qu√£o bem a estrutura do espa√ßo de entrada √© modelada pelo mapa. Esta medida avalia especificamente a discontinuidades no mapa. √â calculada utilizando a dist√¢ncia do BMU (Best Matching Unit) e o Second Best Unit. Se estas duas unidades estiverem perto significa que a topologia foi preservada.
 
*As caracter√≠sticas de qualidade podem ser checadas usando os dados de treinamento* 
 
Refer√™ncia: https://homepage.cs.uri.edu/faculty/hamel/pubs/theses/ms-thesis-breard.pdf


```python
# Cria objeto SOM superf√≠cie 10x10 e 3 features. Sigma √© o raio de influ√™ncia em rela√ß√£o ao BMU
som = MiniSom(10, 10, 3, learning_rate=0.2, sigma=1.2) # learning_rate e sigma  / som = MiniSom(10, 10, 3, learning_rate=0.001, sigma=1.2) 
# som.random_weights_init(data=train_data)
rand = np.random.RandomState(1)
#Inicializa mapa com pesos aleat√≥rios
som.random_weights_init(data=rand.randint(0, 255, (10, 3)))

max_iter = 100000  #Define m√°ximo n√∫mero de itera√ß√£o
q_error = []
t_error = []

#Roda  treinamento da SOM passo-a-passo para plotar erro de quantiza√ß√£o e topografia em cada etapa
for i in range(max_iter):
    rand_i = np.random.randint(len(train_data))
    som.update(train_data[rand_i], som.winner(train_data[rand_i]), i, max_iter)
    q_error.append(som.quantization_error(train_data))
    t_error.append(som.topographic_error(train_data))

plt.plot(np.arange(max_iter), q_error, label='Erro de Quantiza√ß√£o')
plt.plot(np.arange(max_iter), t_error, label='Erro Topogr√°fico')
plt.ylabel('Erro de Quantiza√ß√£o')
plt.xlabel('N√∫mero de Itera√ß√µes')
plt.legend()
plt.show()
```
