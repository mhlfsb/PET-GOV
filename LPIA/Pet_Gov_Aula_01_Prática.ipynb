# Visualização de resultado em tela
Outra importante função dos notebooks (e do Colab) é permitir a rápida visualização de dados para visualização das ciências de dados. 

```python
#esse trecho de código são importações de bibliotecas que vocês verão mais à frente do curso
import numpy as np 
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(100) #gera um array de inteiros aleatórios de 0 a 100.
x = [x for x in range(len(ys))] #define o intervalo de x no gráfico

plt.plot(x, ys, '-') #exibe um gráfico de linhas com as informações de ys e x
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6) #estiliza o gráfico com cores, diemnsão e transparência 

plt.title("Exemplo de visualização de dados randômicos")
plt.show()
```
