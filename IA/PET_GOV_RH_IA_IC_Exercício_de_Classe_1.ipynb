## **Sistema Especialista *Fuzzy* para controle de uma lavadora de roupa industrial**

image.png

<p align="justify">Uma lavadora de roupa ou mquina de lavar roupa  um equipamento industrial ou domstico utilizado para lavar roupas utilizando gua. Consiste, basicamente, em um recipiente (tambor) que se enche de gua onde um detergente em p ou lquido  misturado  roupa a ser lavada (carga). Atravs de agitao mecnica promovida pela rotao do tambor, o detergente entra em contato com a sujidade presente na carga e atravs de uma reao qumica, facilitada pela temperatura, quebra e encapsula a sujidade em pequenas partculas que so carregadas mecanicamente pela gua. Dependendo da quantidade e do tipo de sujidade presente na carga este ciclo precisa ser repetido algumas vezes para que a limpeza seja eficaz. A mquina possui um sensor de carga que  capaz de estimar o volume de carga, e tambm um sensor de sujidade infravermelho capaz de medir o nvel de sujidade e estimar o tipo de sujidade predominante na carga.</p>
 
**Em resumo:**
 
Existem 3 principais variveis de entrada no sistema da lavadora.
1. **Volume de carga:**  o volume de roupa que  carregado na lavadora.
2. **Nvel de sujidade:** Nvel de sujidade presente na carga
3. **Tipo de sujidade:** Proteica, lipdica, hidrato de carbono, etc
 
 
Existem 3 parmetros de funcionamento principais de uma lavadora.
1. **Tempo de lavagem:** Proporcional  quantidade de ciclos de lavagem necessrios para a eficcia da limpeza.
2. **Nvel de gua:** Nvel de gua utilizado no tambor. Este parmetro varia com o volume da carga
3. **Temperatura:** Indica a temperatura do processo e depende do tipo e nvel de sujidade presentes na carga.

**<u>Consideraes para este exerccio</u>**
 
Para facilitar o entendimento, no nosso exemplo, vamos considerar apenas 2 variveis de entrada e 1 varivel de sada. Vamos considerar tambm que o tipo de sujidade  apenas o teor de lipdios (quo oleosa), apesar de existirem diversos tipos de sujidade.

## **Funes de Pertinncia**



Variveis de entrada:
1.  **Tipo de sujidade:** pouco oleosa, mdio oleosa, muito oleosa

#### **Variveis de entrada**

##### Varivel Lingustica: Tipo de sujidade




```python
# Define as variveis fuzzy
# Tipo de sujidade tem intervalo de 0 a 10 (Nvel de sujidade)
tipo_sujidade = np.arange(0, 11, 1)
# Cria as funes de pertinncia da primeira varivel
pouco_oleosa = fuzz.trapmf(tipo_sujidade, [0, 0, 2, 5])
medio_oleosa = fuzz.trimf(tipo_sujidade, [2, 5, 8])
muito_oleosa = fuzz.trapmf(tipo_sujidade, [5, 8, 10, 10])
```

###### Visualizao

```python
fig0, ax0 = plt.subplots(figsize=(15, 5))
ax0.plot(tipo_sujidade, pouco_oleosa, 'b', linewidth=1.5, label='pouco oleosa')
ax0.plot(tipo_sujidade, medio_oleosa, 'g', linewidth=1.5, label='mdio oleosa')
ax0.plot(tipo_sujidade, muito_oleosa, 'r', linewidth=1.5, label='muito oleosa')
ax0.set_title('Tipo de sujidade')
ax0.legend()
```

##### Varivel Lingustica: Nvel de sujidade

```python
# Define as variveis fuzzy
# nvel de sujidade tem intervalo de 0 a 10
quantidade_sujidade  = np.arange(0, 11, 1)

# Cria as funes de pertinncia
pouco_suja = fuzz.trapmf(quantidade_sujidade, [0, 0, 2, 5])
medio_suja = fuzz.trimf(quantidade_sujidade, [2, 5, 8])
muito_suja = fuzz.trapmf(quantidade_sujidade, [5, 8, 10, 10])
```

###### Visualizao

```python
fig1, ax1 = plt.subplots(figsize=(15, 5))
ax1.plot(quantidade_sujidade, pouco_suja, 'b', linewidth=1.5, label='pouco suja')
ax1.plot(quantidade_sujidade, medio_suja, 'g', linewidth=1.5, label='mdio suja')
ax1.plot(quantidade_sujidade, muito_suja, 'r', linewidth=1.5, label='muito suja')
ax1.set_title('Nvel de sujidade')
ax1.legend()
```

#### **Varivel de Sada**

##### Varivel Lingustica: Tempo de lavagem

```python
# Define as variveis fuzzy
# Tempo de lavagem tem intervalo de 20 a 120
tempo_lavagem  = np.arange(20, 121, 1)

# Cria as funes de pertinncia
muito_curto = fuzz.trapmf(tempo_lavagem, [0,20,30, 40])
curto = fuzz.trapmf(tempo_lavagem, [30, 40, 50, 60])
medio = fuzz.trapmf(tempo_lavagem, [50, 60, 80,90])
longo = fuzz.trapmf(tempo_lavagem, [80 ,90, 100, 110])
muito_longo = fuzz.trapmf(tempo_lavagem, [100, 110,120, 120])
```

###### Visualizao

```python
sfig2, ax2 = plt.subplots(figsize=(15, 5))

ax2.plot(tempo_lavagem, muito_curto, 'k', linewidth=1.5, label='muito curto')
ax2.plot(tempo_lavagem, curto, 'y', linewidth=1.5, label='curto')
ax2.plot(tempo_lavagem, medio, 'g', linewidth=1.5, label='mdio')
ax2.plot(tempo_lavagem, longo, 'b', linewidth=1.5, label='longo')
ax2.plot(tempo_lavagem, muito_longo, 'r', linewidth=1.5, label='muito longo')
ax2.set_title('Tempo de lavagem')
ax2.legend()
```

## **Fuzzificao**

Definimos aqui os valores das variveis de entrada do sistema *fuzzy*. 

```python
#TESTE OS CENRIOS AQUI
tipo_sujidade_teste = 3             
nivel_sujidade_teste = 7
```

Precisamos agora ativar as funes de pertinncia *fuzzy*. Os valores exatos da funo de pertinncia *fuzzy* 2 e 6, por exemplo, ainda no existem. Fazemos isso com `fuzz.interp_membership`.

```python
pouco_oleosa_teste = fuzz.interp_membership(tipo_sujidade, pouco_oleosa, tipo_sujidade_teste)
medio_oleosa_teste = fuzz.interp_membership(tipo_sujidade, medio_oleosa, tipo_sujidade_teste)
muito_oleosa_teste = fuzz.interp_membership(tipo_sujidade, muito_oleosa, tipo_sujidade_teste)
print('tipo_sujidade_teste = {:.1f}% pouco oleosa + {:.1f}% medio oleosa + {:.1f}%muito oleosa'.format(pouco_oleosa_teste*100,medio_oleosa_teste*100,muito_oleosa_teste*100))

pouco_suja_teste = fuzz.interp_membership(quantidade_sujidade, pouco_suja, nivel_sujidade_teste)
medio_suja_teste = fuzz.interp_membership(quantidade_sujidade, medio_suja, nivel_sujidade_teste)
muito_suja_teste = fuzz.interp_membership(quantidade_sujidade, muito_suja, nivel_sujidade_teste)
print('nivel_sujidade_teste = {:.1f}% pouco suja + {:.1f}% mdio suja + {:.1f}% muito suja'.format(pouco_suja_teste*100,medio_suja_teste*100,muito_suja_teste*100))
```

## **Processamento Fuzzy**

Agora aplicamos nossas regras *fuzzy*. Devemos usar `np.fmin` para lgica `AND` e `np.fmax` para lgica `OR`.

```python
#1. Se a nvel de sujidade  pouco suja E tipo de sujidade  pouco oleosa ...
active_rule1 = np.fmin(pouco_oleosa_teste, pouco_suja_teste) #Lgica E (AND)
# ...ENTO tempo de lavagem  muito curto
muito_curto_atv = np.fmin(active_rule1, muito_curto) 
#Ativa vrivel de sada "muito curto"

#2. Se nvel de sujidade  mdio suja E tipo de sujidade  pouco oleosa ...
active_rule2 = np.fmin(pouco_oleosa_teste, medio_suja_teste)
# ENTO tempo de lavagem  curto
curto_atv = np.fmin(active_rule2, curto)
# Ativa vrivel de sada "curto"

#3. Se nvel de sujidade  muito suja E tipo de sujidade  pouco oleosa ...
active_rule3 = np.fmin(pouco_oleosa_teste, muito_suja_teste)
#4. Se nvel de sujidade  pouco suja E tipo de sujidade  mdio oleosa ... 
active_rule4 = np.fmin(medio_oleosa_teste, pouco_suja_teste)
#5. Se nvel de sujidade  mdio suja E tipo de sujidade  mdio oleosa ...
active_rule5 = np.fmin(medio_oleosa_teste, medio_suja_teste)
#6. Se nvel de sujidade  muito suja E tipo de sujidade  mdio oleosa ...
active_rule6 = np.fmin(medio_oleosa_teste, muito_suja_teste)
# Agrega todas as regras com lgica OU (max) pois a sada para 
# todas elas  "mdio"
medio_atv = np.fmax(active_rule3, active_rule4)
medio_atv = np.fmax(active_rule5, medio_atv)
medio_atv = np.fmax(active_rule6, medio_atv)
# ... ENTO  tempo de lavagem  mdio
medio_atv = np.fmin(medio_atv, medio)
# Ativa vrivel de sada "mdio"

#7. Se nvel de sujidade  pouco suja E tipo de sujidade  muito oleosa ...
active_rule7 = np.fmin(muito_oleosa_teste, pouco_suja_teste)
#8. Se nvel de sujidade  mdio suja E tipo de sujidade  muito oleosa ...
active_rule8 = np.fmin(muito_oleosa_teste, medio_suja_teste)
# Agrega todas as regras com lgica OU (max) pois a sada para 
# todas elas  "longo"
longo_atv = np.fmax(active_rule7,active_rule8)
# ENTO tempo de lavagem  longo
longo_atv = np.fmin(longo_atv,longo)
# Ativa vrivel de sada "longo"

#9. Se nvel de sujidade  muito suja E tipo de sujidade  muito oleosa ...
active_rule9 = np.fmin(muito_oleosa_teste, muito_suja_teste)
# ENTO tempo de lavagem  muito longo
muito_longo_atv = np.fmin(active_rule9,muito_longo)
# Ativa vrivel de sada "muito longo"

# Aqui ns agregamos todas as funes de pertinncia de sada 
aggregated = np.fmax(muito_curto_atv,curto_atv)
aggregated = np.fmax(medio_atv,aggregated)
aggregated = np.fmax(longo_atv,aggregated)
aggregated = np.fmax(muito_longo_atv,aggregated)
```

##### Visualizao

```python
out0 = np.zeros_like(tempo_lavagem)
fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.fill_between(tempo_lavagem, out0, muito_curto_atv, facecolor='k', alpha=0.7)
ax0.plot(tempo_lavagem, muito_curto_atv, 'b', linewidth=0.5, linestyle='--', )

ax0.fill_between(tempo_lavagem, out0, curto_atv, facecolor='y', alpha=0.7)
ax0.plot(tempo_lavagem, curto_atv, 'g', linewidth=0.5, linestyle='--')

ax0.fill_between(tempo_lavagem, out0, medio_atv, facecolor='g', alpha=0.7)
ax0.plot(tempo_lavagem, medio_atv, 'g', linewidth=0.5, linestyle='--')

ax0.fill_between(tempo_lavagem, out0, longo_atv, facecolor='b', alpha=0.7)
ax0.plot(tempo_lavagem, longo_atv, 'g', linewidth=0.5, linestyle='--')

ax0.fill_between(tempo_lavagem, out0, muito_longo_atv, facecolor='r', alpha=0.7)
ax0.plot(tempo_lavagem, muito_longo_atv, 'g', linewidth=0.5, linestyle='--')


# Turn off top/right axes
for ax in (ax0,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

## **Defuzzificao** 

**Vamos agora realizar a defuzzificao e calcular a sada do sistema especialista fuzzy**

```python
# Calcula o resultado aps a defuzzificao
tempo_lavagem_saida = fuzz.defuzz(tempo_lavagem, aggregated, 'centroid')
print('Para \nTipo de sujidade: {}\nnvel de sujidade: {}'.format(tipo_sujidade_teste,
                                                          nivel_sujidade_teste))
print('\n\nTempo de lavagem adequado  {} minutos'.format(int(tempo_lavagem_saida)))
```

```python
saida_ativada = fuzz.interp_membership(tempo_lavagem, aggregated, tempo_lavagem_saida)  

fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.plot(tempo_lavagem, muito_curto_atv, 'b', linewidth=0.5, linestyle='--', )
ax0.plot(tempo_lavagem, curto_atv, 'g', linewidth=0.5, linestyle='--')
ax0.plot(tempo_lavagem, medio_atv, 'r', linewidth=0.5, linestyle='--')
ax0.plot(tempo_lavagem, longo_atv, 'r', linewidth=0.5, linestyle='--')
ax0.plot(tempo_lavagem, muito_longo_atv, 'r', linewidth=0.5, linestyle='--')
ax0.fill_between(tempo_lavagem, out0, aggregated, facecolor='Orange', alpha=0.7)
ax0.plot([tempo_lavagem_saida, tempo_lavagem_saida], [0, saida_ativada], 'k', linewidth=1.5, alpha=0.9)
ax0.set_title('Funo de pertinncia agregada de sada e resultado (linha)')

# Turn off top/right axes
for ax in (ax0,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

```python
tabela_tempo_lavagem = pd.DataFrame()
tempo_lavagem_saida = []
for tipo in range(0,11):
  tempo_lavagem_saida = []
  for quant in range(0,11):
    pouco_oleosa_teste = fuzz.interp_membership(tipo_sujidade, pouco_oleosa, tipo)
    medio_oleosa_teste = fuzz.interp_membership(tipo_sujidade, medio_oleosa, tipo)
    muito_oleosa_teste = fuzz.interp_membership(tipo_sujidade, muito_oleosa, tipo)

    pouco_suja_teste = fuzz.interp_membership(quantidade_sujidade, pouco_suja, quant)
    medio_suja_teste = fuzz.interp_membership(quantidade_sujidade, medio_suja, quant)
    muito_suja_teste = fuzz.interp_membership(quantidade_sujidade, muito_suja, quant)

    #1. Se nvel de sujidade  pouco suja E tipo de sujidade  pouco oleosa ...
    active_rule1 = np.fmin(pouco_oleosa_teste, pouco_suja_teste) #Lgica E (AND)
    # ...ENTO tempo de lavagem  muito curto
    muito_curto_atv = np.fmin(active_rule1, muito_curto) 
    #Ativa vrivel de sada "muito curto"

    #2. Se nvel de sujidade  mdio suja E tipo de sujidade  pouco oleosa ...
    active_rule2 = np.fmin(pouco_oleosa_teste, medio_suja_teste)
    # ENTO tempo de lavagem  curto
    curto_atv = np.fmin(active_rule2, curto)
    # Ativa vrivel de sada "curto"

    #3. Se nvel de sujidade  muito suja E tipo de sujidade  pouco oleosa ...
    active_rule3 = np.fmin(pouco_oleosa_teste, muito_suja_teste)
    #4. Se nvel de sujidade  pouco suja E tipo de sujidade  mdio oleosa ... 
    active_rule4 = np.fmin(medio_oleosa_teste, pouco_suja_teste)
    #5. Se nvel de sujidade  mdio suja E tipo de sujidade  mdio oleosa ...
    active_rule5 = np.fmin(medio_oleosa_teste, medio_suja_teste)
    #6. Se nvel de sujidade  muito suja E tipo de sujidade  mdio oleosa ...
    active_rule6 = np.fmin(medio_oleosa_teste, muito_suja_teste)
    # Agrega todas as regras com lgica OU (max) pois a sada para 
    # todas elas  "mdio"
    medio_atv = np.fmax(active_rule3, active_rule4)
    medio_atv = np.fmax(active_rule5, medio_atv)
    medio_atv = np.fmax(active_rule6, medio_atv)
    # ... ENTO  tempo de lavagem  mdio
    medio_atv = np.fmin(medio_atv, medio)
    # Ativa vrivel de sada "mdio"

    #7. Se nvel de sujidade  pouco suja E tipo de sujidade  muito oleosa ...
    active_rule7 = np.fmin(muito_oleosa_teste, pouco_suja_teste)
    #8. Se nvel de sujidade  mdio suja E tipo de sujidade  muito oleosa ...
    active_rule8 = np.fmin(muito_oleosa_teste, medio_suja_teste)
    # Agrega todas as regras com lgica OU (max) pois a sada para 
    # todas elas  "longo"
    longo_atv = np.fmax(active_rule7,active_rule8)
    # ENTO tempo de lavagem  longo
    longo_atv = np.fmin(longo_atv,longo)
    # Ativa vrivel de sada "longo"

    #9. Se nvel de sujidade  muito suja E tipo de sujidade  muito oleosa ...
    active_rule9 = np.fmin(muito_oleosa_teste, muito_suja_teste)
    # ENTO tempo de lavagem  muito longo
    muito_longo_atv = np.fmin(active_rule9,muito_longo)
    # Ativa vrivel de sada "muito longo"

    # Aqui ns agregamos todas as funes de pertinncia de sada 
    aggregated = np.fmax(muito_curto_atv,curto_atv)
    aggregated = np.fmax(medio_atv,aggregated)
    aggregated = np.fmax(longo_atv,aggregated)
    aggregated = np.fmax(muito_longo_atv,aggregated)
    tempo_lavagem_saida.append(fuzz.defuzz(tempo_lavagem, aggregated, 'centroid'))
  serie = pd.Series(tempo_lavagem_saida).astype(int)
  saida_pandas = pd.DataFrame({str(tipo):serie})
  tabela_tempo_lavagem = pd.concat([tabela_tempo_lavagem,saida_pandas], axis=1)
```
