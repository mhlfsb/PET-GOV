## **Sistema Especialista *Fuzzy* para controle de uma lavadora de roupa industrial**

image.png

<p align="justify">Uma lavadora de roupa ou máquina de lavar roupa é um equipamento industrial ou doméstico utilizado para lavar roupas utilizando água. Consiste, basicamente, em um recipiente (tambor) que se enche de água onde um detergente em pó ou líquido é misturado à roupa a ser lavada (carga). Através de agitação mecânica promovida pela rotação do tambor, o detergente entra em contato com a sujidade presente na carga e através de uma reação química, facilitada pela temperatura, quebra e encapsula a sujidade em pequenas partículas que são carregadas mecanicamente pela água. Dependendo da quantidade e do tipo de sujidade presente na carga este ciclo precisa ser repetido algumas vezes para que a limpeza seja eficaz. A máquina possui um sensor de carga que é capaz de estimar o volume de carga, e também um sensor de sujidade infravermelho capaz de medir o nível de sujidade e estimar o tipo de sujidade predominante na carga.</p>
 
**Em resumo:**
 
Existem 3 principais variáveis de entrada no sistema da lavadora.
1. **Volume de carga:** É o volume de roupa que é carregado na lavadora.
2. **Nível de sujidade:** Nível de sujidade presente na carga
3. **Tipo de sujidade:** Proteica, lipídica, hidrato de carbono, etc
 
 
Existem 3 parâmetros de funcionamento principais de uma lavadora.
1. **Tempo de lavagem:** Proporcional à quantidade de ciclos de lavagem necessários para a eficácia da limpeza.
2. **Nível de água:** Nível de água utilizado no tambor. Este parâmetro varia com o volume da carga
3. **Temperatura:** Indica a temperatura do processo e depende do tipo e nível de sujidade presentes na carga.

#### **Variáveis de entrada**

##### Variável Linguística: Tipo de sujidade




```python
# Define as variáveis fuzzy
# Tipo de sujidade tem intervalo de 0 a 10 (Nível de sujidade)
tipo_sujidade = np.arange(0, 11, 1)
# Cria as funções de pertinência da primeira variável
pouco_oleosa = fuzz.trapmf(tipo_sujidade, [0, 0, 2, 5])
medio_oleosa = fuzz.trimf(tipo_sujidade, [2, 5, 8])
muito_oleosa = fuzz.trapmf(tipo_sujidade, [5, 8, 10, 10])
```

###### Visualização

##### Variável Linguística: Nível de sujidade

```python
# Define as variáveis fuzzy
# nível de sujidade tem intervalo de 0 a 10
quantidade_sujidade  = np.arange(0, 11, 1)

# Cria as funções de pertinência
pouco_suja = fuzz.trapmf(quantidade_sujidade, [0, 0, 2, 5])
medio_suja = fuzz.trimf(quantidade_sujidade, [2, 5, 8])
muito_suja = fuzz.trapmf(quantidade_sujidade, [5, 8, 10, 10])
```

###### Visualização

```python
fig1, ax1 = plt.subplots(figsize=(15, 5))
ax1.plot(quantidade_sujidade, pouco_suja, 'b', linewidth=1.5, label='pouco suja')
ax1.plot(quantidade_sujidade, medio_suja, 'g', linewidth=1.5, label='médio suja')
ax1.plot(quantidade_sujidade, muito_suja, 'r', linewidth=1.5, label='muito suja')
ax1.set_title('Nível de sujidade')
ax1.legend()
```

#### **Variável de Saída**

##### Variável Linguística: Tempo de lavagem

```python
# Define as variáveis fuzzy
# Tempo de lavagem tem intervalo de 20 a 120
tempo_lavagem  = np.arange(20, 121, 1)

# Cria as funções de pertinência
muito_curto = fuzz.trapmf(tempo_lavagem, [0,20,30, 40])
curto = fuzz.trapmf(tempo_lavagem, [30, 40, 50, 60])
medio = fuzz.trapmf(tempo_lavagem, [50, 60, 80,90])
longo = fuzz.trapmf(tempo_lavagem, [80 ,90, 100, 110])
muito_longo = fuzz.trapmf(tempo_lavagem, [100, 110,120, 120])
```

## **Fuzzificação**

Definimos aqui os valores das variáveis de entrada do sistema *fuzzy*. 

```python
#TESTE OS CENÁRIOS AQUI
tipo_sujidade_teste = 3             
nivel_sujidade_teste = 7
```

Precisamos agora ativar as funções de pertinência *fuzzy*. Os valores exatos da função de pertinência *fuzzy* 2 e 6, por exemplo, ainda não existem. Fazemos isso com `fuzz.interp_membership`.

```python
pouco_oleosa_teste = fuzz.interp_membership(tipo_sujidade, pouco_oleosa, tipo_sujidade_teste)
medio_oleosa_teste = fuzz.interp_membership(tipo_sujidade, medio_oleosa, tipo_sujidade_teste)
muito_oleosa_teste = fuzz.interp_membership(tipo_sujidade, muito_oleosa, tipo_sujidade_teste)
print('tipo_sujidade_teste = {:.1f}% pouco oleosa + {:.1f}% medio oleosa + {:.1f}%muito oleosa'.format(pouco_oleosa_teste*100,medio_oleosa_teste*100,muito_oleosa_teste*100))

pouco_suja_teste = fuzz.interp_membership(quantidade_sujidade, pouco_suja, nivel_sujidade_teste)
medio_suja_teste = fuzz.interp_membership(quantidade_sujidade, medio_suja, nivel_sujidade_teste)
muito_suja_teste = fuzz.interp_membership(quantidade_sujidade, muito_suja, nivel_sujidade_teste)
print('nivel_sujidade_teste = {:.1f}% pouco suja + {:.1f}% médio suja + {:.1f}% muito suja'.format(pouco_suja_teste*100,medio_suja_teste*100,muito_suja_teste*100))
```

## **Processamento Fuzzy**

Agora aplicamos nossas regras *fuzzy*. Devemos usar `np.fmin` para lógica `AND` e `np.fmax` para lógica `OR`.

```python
#1. Se a nível de sujidade é pouco suja E tipo de sujidade é pouco oleosa ...
active_rule1 = np.fmin(pouco_oleosa_teste, pouco_suja_teste) #Lógica E (AND)
# ...ENTÃO tempo de lavagem é muito curto
muito_curto_atv = np.fmin(active_rule1, muito_curto) 
#Ativa váriável de saída "muito curto"

#2. Se nível de sujidade é médio suja E tipo de sujidade é pouco oleosa ...
active_rule2 = np.fmin(pouco_oleosa_teste, medio_suja_teste)
# ENTÃO tempo de lavagem é curto
curto_atv = np.fmin(active_rule2, curto)
# Ativa váriável de saída "curto"

#3. Se nível de sujidade é muito suja E tipo de sujidade é pouco oleosa ...
active_rule3 = np.fmin(pouco_oleosa_teste, muito_suja_teste)
#4. Se nível de sujidade é pouco suja E tipo de sujidade é médio oleosa ... 
active_rule4 = np.fmin(medio_oleosa_teste, pouco_suja_teste)
#5. Se nível de sujidade é médio suja E tipo de sujidade é médio oleosa ...
active_rule5 = np.fmin(medio_oleosa_teste, medio_suja_teste)
#6. Se nível de sujidade é muito suja E tipo de sujidade é médio oleosa ...
active_rule6 = np.fmin(medio_oleosa_teste, muito_suja_teste)
# Agrega todas as regras com lógica OU (max) pois a saída para 
# todas elas é "médio"
medio_atv = np.fmax(active_rule3, active_rule4)
medio_atv = np.fmax(active_rule5, medio_atv)
medio_atv = np.fmax(active_rule6, medio_atv)
# ... ENTÃO  tempo de lavagem é médio
medio_atv = np.fmin(medio_atv, medio)
# Ativa váriável de saída "médio"

#7. Se nível de sujidade é pouco suja E tipo de sujidade é muito oleosa ...
active_rule7 = np.fmin(muito_oleosa_teste, pouco_suja_teste)
#8. Se nível de sujidade é médio suja E tipo de sujidade é muito oleosa ...
active_rule8 = np.fmin(muito_oleosa_teste, medio_suja_teste)
# Agrega todas as regras com lógica OU (max) pois a saída para 
# todas elas é "longo"
longo_atv = np.fmax(active_rule7,active_rule8)
# ENTÃO tempo de lavagem é longo
longo_atv = np.fmin(longo_atv,longo)
# Ativa váriável de saída "longo"

#9. Se nível de sujidade é muito suja E tipo de sujidade é muito oleosa ...
active_rule9 = np.fmin(muito_oleosa_teste, muito_suja_teste)
# ENTÃO tempo de lavagem é muito longo
muito_longo_atv = np.fmin(active_rule9,muito_longo)
# Ativa váriável de saída "muito longo"

# Aqui nós agregamos todas as funções de pertinência de saída 
aggregated = np.fmax(muito_curto_atv,curto_atv)
aggregated = np.fmax(medio_atv,aggregated)
aggregated = np.fmax(longo_atv,aggregated)
aggregated = np.fmax(muito_longo_atv,aggregated)
```

##### Visualização

```python
out0 = np.zeros_like(tempo_lavagem)
fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.fill_between(tempo_lavagem, out0, muito_curto_atv, facecolor='k', alpha=0.7)
ax0.plot(tempo_lavagem, muito_curto_atv, 'b', linewidth=0.5, linestyle='--', )

ax0.fill_between(tempo_lavagem, out0, curto_atv, facecolor='y', alpha=0.7)
ax0.plot(tempo_lavagem, curto_atv, 'g', linewidth=0.5, linestyle='--')

ax0.fill_between(tempo_lavagem, out0, medio_atv, facecolor='g', alpha=0.7)
ax0.plot(tempo_lavagem, medio_atv, 'g', linewidth=0.5, linestyle='--')

ax0.fill_between(tempo_lavagem, out0, longo_atv, facecolor='b', alpha=0.7)
ax0.plot(tempo_lavagem, longo_atv, 'g', linewidth=0.5, linestyle='--')

ax0.fill_between(tempo_lavagem, out0, muito_longo_atv, facecolor='r', alpha=0.7)
ax0.plot(tempo_lavagem, muito_longo_atv, 'g', linewidth=0.5, linestyle='--')


# Turn off top/right axes
for ax in (ax0,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

## **Defuzzificação** 

**Vamos agora realizar a defuzzificação e calcular a saída do sistema especialista fuzzy**

```python
# Calcula o resultado após a defuzzificação
tempo_lavagem_saida = fuzz.defuzz(tempo_lavagem, aggregated, 'centroid')
print('Para \nTipo de sujidade: {}\nnível de sujidade: {}'.format(tipo_sujidade_teste,
                                                          nivel_sujidade_teste))
print('\n\nTempo de lavagem adequado é {} minutos'.format(int(tempo_lavagem_saida)))
```

```python
saida_ativada = fuzz.interp_membership(tempo_lavagem, aggregated, tempo_lavagem_saida)  

fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.plot(tempo_lavagem, muito_curto_atv, 'b', linewidth=0.5, linestyle='--', )
ax0.plot(tempo_lavagem, curto_atv, 'g', linewidth=0.5, linestyle='--')
ax0.plot(tempo_lavagem, medio_atv, 'r', linewidth=0.5, linestyle='--')
ax0.plot(tempo_lavagem, longo_atv, 'r', linewidth=0.5, linestyle='--')
ax0.plot(tempo_lavagem, muito_longo_atv, 'r', linewidth=0.5, linestyle='--')
ax0.fill_between(tempo_lavagem, out0, aggregated, facecolor='Orange', alpha=0.7)
ax0.plot([tempo_lavagem_saida, tempo_lavagem_saida], [0, saida_ativada], 'k', linewidth=1.5, alpha=0.9)
ax0.set_title('Função de pertinência agregada de saída e resultado (linha)')

# Turn off top/right axes
for ax in (ax0,):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```

### **Extra**

Podemos usar os sistema especialista *fuzzy* de várias maneiras.

Porém existe uma maneira muito simples de se usar um sistema especialista sem a necessidade de utilizar um computador! Se o número de variáveis de entrada for menor que 2 podemos transformar o conhecimento do especialista em uma tabela. Por quê não?

Desta forma, uma pessoa sem conhecimento especialista pode facilmente fazer uso deste conhecimento, simplesmente consultando a tabela.


```python
tabela_tempo_lavagem = pd.DataFrame()
tempo_lavagem_saida = []
for tipo in range(0,11):
  tempo_lavagem_saida = []
  for quant in range(0,11):
    pouco_oleosa_teste = fuzz.interp_membership(tipo_sujidade, pouco_oleosa, tipo)
    medio_oleosa_teste = fuzz.interp_membership(tipo_sujidade, medio_oleosa, tipo)
    muito_oleosa_teste = fuzz.interp_membership(tipo_sujidade, muito_oleosa, tipo)

    pouco_suja_teste = fuzz.interp_membership(quantidade_sujidade, pouco_suja, quant)
    medio_suja_teste = fuzz.interp_membership(quantidade_sujidade, medio_suja, quant)
    muito_suja_teste = fuzz.interp_membership(quantidade_sujidade, muito_suja, quant)

    #1. Se nível de sujidade é pouco suja E tipo de sujidade é pouco oleosa ...
    active_rule1 = np.fmin(pouco_oleosa_teste, pouco_suja_teste) #Lógica E (AND)
    # ...ENTÃO tempo de lavagem é muito curto
    muito_curto_atv = np.fmin(active_rule1, muito_curto) 
    #Ativa váriável de saída "muito curto"

    #2. Se nível de sujidade é médio suja E tipo de sujidade é pouco oleosa ...
    active_rule2 = np.fmin(pouco_oleosa_teste, medio_suja_teste)
    # ENTÃO tempo de lavagem é curto
    curto_atv = np.fmin(active_rule2, curto)
    # Ativa váriável de saída "curto"

    #3. Se nível de sujidade é muito suja E tipo de sujidade é pouco oleosa ...
    active_rule3 = np.fmin(pouco_oleosa_teste, muito_suja_teste)
    #4. Se nível de sujidade é pouco suja E tipo de sujidade é médio oleosa ... 
    active_rule4 = np.fmin(medio_oleosa_teste, pouco_suja_teste)
    #5. Se nível de sujidade é médio suja E tipo de sujidade é médio oleosa ...
    active_rule5 = np.fmin(medio_oleosa_teste, medio_suja_teste)
    #6. Se nível de sujidade é muito suja E tipo de sujidade é médio oleosa ...
    active_rule6 = np.fmin(medio_oleosa_teste, muito_suja_teste)
    # Agrega todas as regras com lógica OU (max) pois a saída para 
    # todas elas é "médio"
    medio_atv = np.fmax(active_rule3, active_rule4)
    medio_atv = np.fmax(active_rule5, medio_atv)
    medio_atv = np.fmax(active_rule6, medio_atv)
    # ... ENTÃO  tempo de lavagem é médio
    medio_atv = np.fmin(medio_atv, medio)
    # Ativa váriável de saída "médio"

    #7. Se nível de sujidade é pouco suja E tipo de sujidade é muito oleosa ...
    active_rule7 = np.fmin(muito_oleosa_teste, pouco_suja_teste)
    #8. Se nível de sujidade é médio suja E tipo de sujidade é muito oleosa ...
    active_rule8 = np.fmin(muito_oleosa_teste, medio_suja_teste)
    # Agrega todas as regras com lógica OU (max) pois a saída para 
    # todas elas é "longo"
    longo_atv = np.fmax(active_rule7,active_rule8)
    # ENTÃO tempo de lavagem é longo
    longo_atv = np.fmin(longo_atv,longo)
    # Ativa váriável de saída "longo"

    #9. Se nível de sujidade é muito suja E tipo de sujidade é muito oleosa ...
    active_rule9 = np.fmin(muito_oleosa_teste, muito_suja_teste)
    # ENTÃO tempo de lavagem é muito longo
    muito_longo_atv = np.fmin(active_rule9,muito_longo)
    # Ativa váriável de saída "muito longo"

    # Aqui nós agregamos todas as funções de pertinência de saída 
    aggregated = np.fmax(muito_curto_atv,curto_atv)
    aggregated = np.fmax(medio_atv,aggregated)
    aggregated = np.fmax(longo_atv,aggregated)
    aggregated = np.fmax(muito_longo_atv,aggregated)
    tempo_lavagem_saida.append(fuzz.defuzz(tempo_lavagem, aggregated, 'centroid'))
  serie = pd.Series(tempo_lavagem_saida).astype(int)
  saida_pandas = pd.DataFrame({str(tipo):serie})
  tabela_tempo_lavagem = pd.concat([tabela_tempo_lavagem,saida_pandas], axis=1)
```
