### Redundância nos dados

```python
# Primeiro criamos um DataFrame com dados redundantes
df = pd.DataFrame({
    "Nome": ["João", "Maria", "José", "Silvio"],
    "Gênero": ["Masculino", "Feminino", "Masculino", "Masculino"],
    "UF": ["PE", "PE", "PB", "RN"]
  }
)
df.head()
```

### Linhas duplicadas

```python
# Criando um Dataframe com linhas duplicados
df = pd.DataFrame({'Nome': ['João', 'Maria', 'João', 'Silvio'], 'Gênero': ['Masculino', 'Feminino', 'Masculino', 'Masculino'], 'UF': ['PE', 'PE', 'PE', 'RN']})
df.head()
```

### Codificação

```python
# Criando um Dataframe com dados para serem codificados
df = pd.DataFrame({
    'Veículo': ['Fiat', 'Ford', 'Fiat', 'Jeep', 'Renault','Peugeot','Toyota','VW'], 
    'Área': ['Capital', 'Interior', 'RMR', 'Capital', 'Capital', 'RMR', 'RMR','Interior'], 
    'Consumo': [10, 15, 8, 6, 6, 6, 9, 5], 
    'Suspeito': ['Sim', 'Sim', 'Não', 'Não', 'Sim', 'Não', 'Sim', 'Não']})
df.head(10)
```

```python
df.describe()
```

```python
lista_tipos_veiculos = df['Veículo'].unique().tolist()
print(lista_tipos_veiculos) #imprimindo os tipos únicos da coluna Veículo
i = 1
for veiculo in lista_tipos_veiculos:
  df.loc[df['Veículo'] == veiculo, 'Veículo'] = int(i)
  i += 1
df
```

```python
df.loc[df['Veículo'] == 'Fiat', 'Veículo'] = 0
df.loc[df['Veículo'] == 'Ford', 'Veículo'] = 1
df.loc[df['Veículo'] == 'Jeep', 'Veículo'] = 2

df.loc[df['Área'] == 'Capital', 'Área'] = 0
df.loc[df['Área'] == 'Interior', 'Área'] = 1
df.loc[df['Área'] == 'RMR', 'Área'] = 2

df.loc[df['Suspeito'] == 'Sim', 'Suspeito'] = 1
df.loc[df['Suspeito'] == 'Não', 'Suspeito'] = 0

df["Veículo"] = pd.to_numeric(df["Veículo"]); 
df["Área"] = pd.to_numeric(df["Área"]); 
df["Suspeito"] = pd.to_numeric(df["Suspeito"]); 

print(df.head(10))
print(df.info())
df.describe() #método/função
df.shape #atributo
```

**Data Dictionary - Titanic:**

* Age = Idade do passageiro
* SibSp = Número de irmãos/irmãs ou maridos/esposas abordos do Titanic
* Parch = Número de pais (pai e mãe) ou filhos abordos do Titanic
* Ticket = Número do ticket (passagem)
* Fare = Valor da tarifa paga pelo passageiro

## Vamos remover colunas indesejadas

```python
# Vamos criar uma lista contendo as colunas indesejadas
colunas_indesejadas = ['PassengerId', 'Name', 'Ticket', 'Cabin', 'SibSp', 'Parch']
print(f'Removendo as colunas indesejadas: {colunas_indesejadas}')

print(f'Data set antes da remoção das colunas: {df.shape[0]} linhas e {df.shape[1]} colunas')

# Removendo as colunas indesejadas.
# O parâmetro 'labels' espera receber uma lista com os nomes das colunas que devem ser removidas (dropadas).
# O parâmetro 'axis' espera receber 0 (ou 'index') para remover linhas ou 1 ('columns') para remover colunas.
# No nosso caso, iremos remover as colunas (axis=1 ou axis='columns')
df = df.drop(labels=colunas_indesejadas, axis=1)

print(f'Data set após a remoção das colunas: {df.shape[0]} linhas e {df.shape[1]} colunas\n')
df.head(10)
```

## Removendo linhas com valores vazios (NaN)

```python
print('Número de linhas vazias para cada coluna (NaN):')
df.isna().sum()
```

```python
# Visualizando uma amostra aleatória de 3 linhas com valores vazios para 'Age'
df[df['Age'].isna()].sample(5)
```

```python
# Visualizando as duas linhas com valores vazios para 'Embarked'
df[df['Embarked'].isna()]
```

```python
# Removendo as linhas com valores vazios (NaN)
# Caso queira remover as colunas que possuem valores vazios, invés de axis=0, usar axis=1
df = df.dropna(axis=0)

print(f'Data set após a remoção de linhas com valores vazios (NaN): {df.shape[0]} linhas e {df.shape[1]} colunas\n')
df.head(10)
```

```python
print('Número de linhas vazias para cada coluna (NaN):')
df.isna().sum()
```

```python
# Usando o 'loc' para localizar as linhas através de um filtro, 
# e selecionar uma coluna específica da coluna e atribuir um novo valor para ela.
df.loc[df['Sex'] == 'male', 'Sex'] = 0
df.loc[df['Sex'] == 'female', 'Sex'] = 1

df.loc[df['Embarked'] == 'C', 'Embarked'] = 0
df.loc[df['Embarked'] == 'Q', 'Embarked'] = 1
df.loc[df['Embarked'] == 'S', 'Embarked'] = 2

print(f'Data set após a codificação dos valores categóricos: {df.shape[0]} linhas e {df.shape[1]} colunas\n')
df.head()
```
